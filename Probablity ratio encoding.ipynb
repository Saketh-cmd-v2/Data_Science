```python
##### Load titanic dataset for demo
df = pd.read_csv('titanic.csv', usecols = ['Cabin', 'Survived'])
df.head()
```

```python
#### Replace the missing values with a new category called missing
df['Cabin'].fillna('Missing', inplace = True)
df.head(20)
```

```python
#### Replace the cabin values only with the first letter
df['Cabin'] = df['Cabin'].astype(str).str[0]
df.head()
```

```python
#### Now find probalblity of survival
prob_df = df.groupby(['Cabin'])['Survived'].mean()
prob_df
```

```python
#### Put this in a dataframe
dataframe = pd.DataFrame(prob_df)
dataframe.head()
```

```python
#### Similarly calculate prob of death
dataframe['Died'] = 1-dataframe['Survived']
dataframe.head()
```

```python
#### Divide prob of survived by prob of death to get prob ratio
dataframe['ProbablityRatio'] = dataframe['Survived']/dataframe['Died']
dataframe.head()
```

```python
##### Now convert it to a dictionary(ProbablityRatio)
prob_encoded = dataframe['ProbablityRatio'].to_dict()
prob_encoded
```

```python
#### Map prob_encoded to the the categorical data to be handled
df['Cabin_handled'] = df['Cabin'].map(prob_encoded)
df.head()
#### Now we can drop the initial cabin 
```
