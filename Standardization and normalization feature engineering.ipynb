```python
import scipy.stats as stat
import pylab
```

```python
#### Gaussian transformations
## Function to check weather the points form a normal distribution or not
def plot_values(df,feature):
    plt.figure(figsize =(10,10))
    plt.subplot(1,2,1)
    df[feature].hist(bins=20)
    plt.subplot(1,2,2)
    stat.probplot(df[feature],dist='norm', plot=pylab )
    
plot_values(df,'Pclass')
```

```python
#### Applying lograthemic transform
df['pclass_log'] = np.log(df['Pclass'])
```

```python
## Not very good as the right graph has many points outside the red line suggesting that after log transform the distribution
## is not normal distribution.
plot_values(df,'Pclass')
```

```python
##Reciprocal transformation
df['pclass_reciprocal'] = 1/df['Pclass']
plot_values(df,'pclass_reciprocal')
```

```python
#### Square root transformation
df['pclass_sq_root'] = df['Pclass']**(1/2)
plot_values(df, 'pclass_sq_root')
```

```python
## Exponential transform
df['p_exp'] = df.Pclass**(1/1.2)
plot_values(df,'p_exp')

```

```python
## Boxcox tarnsformation
#present in the modulescipt.stats.boxcox
import scipy.stats as stat
df['Age_boxcox'], parms = stat.boxcox(df['Age']) 

print(parms)
```

```python
plot_values(df,'Age_boxcox')
```

```python
## Boxcox trans for pclass
df['pclass_boxcox'], parms = stat.boxcox(df['Pclass'])
plot_values(df,'pclass_boxcox')
```

```python
## plot foe fare
plot_values(df, 'Fare')
```

```python
## log to fare
df['fare_log'] = np.log1p(df['Fare'])
plot_values(df,'fare_log')
```

```python
## boxcox for fare
df['fare_boxcox'], parms = stat.boxcox(df['Fare']+1)
```

```python
plot_values(df, 'fare_boxcox')
```

```python
## exponential for fare
df['fare_exp'] = df.Fare**(1/1.2)
plot_values(df,'fare_exp')
```

```python
## Sqrt for fare

df['fare_sqrt'] = df.Fare**(1/2)
plot_values(df,'fare_sqrt')
```

```python
## Reciprocal for fare
df['reci_fare'] = 1/(df['Fare']+1)
plot_values(df, 'reci_fare')
```
